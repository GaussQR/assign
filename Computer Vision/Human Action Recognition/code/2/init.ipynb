```
%matplotlib inline
import cv2
import math
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from keras.preprocessing import image
from glob import glob
from tqdm import tqdm
```

```
base_path = "ucfTrainTestlist/"
f = open(base_path + "trainlist01.txt", "r")
temp = f.read()
train = pd.DataFrame()
train['video_name'] = temp.split('\n')
train = train[:-1]
train.head()
```

```
train_video_tag = []
for i in range(train.shape[0]):
    train_video_tag.append(train['video_name'][i].split('/')[0])
train['tag'] = train_video_tag
```

```
for i in tqdm(range(train.shape[0])):
    videoFile = train['video_name'][i]
    path = 'UCF-101/' + train['tag'][i] + '/' + videoFile.split(' ')[0].split('/')[1]
    cap = cv2.VideoCapture(path)
    frameRate = 2*cap.get(5)
    count = 0
    while(cap.isOpened()):
        frameId = cap.get(1)
        ret, frame = cap.read()
        if ret != True: break
        if frameId % frameRate == 1:
            filename ='train_01/' + videoFile.split('/')[1].split(' ')[0] +"_frame%d.jpg" % count
            count += 1
            cv2.imwrite(filename, frame)
    cap.release()
```
